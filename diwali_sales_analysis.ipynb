# Explotary Data Analysis


### Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Updated code: Using hue for Gender and assigning a palette
ax = sns.countplot(x='Gender', data=df, hue='Gender', palette=['orange', 'blue'], legend=False)

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()


```

```python
# plotting a bar chart for gender vs total amaount 
sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data= sales_gen)
```

# From above graph we can see that most of the buyer are females and even the purchasing power of females are greater than men 


## Age


```python
ax = sns.countplot(data = df, x= 'Age Group' , hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (15, 5)})

# Normalize the 'Amount' for color mapping
norm = plt.Normalize(sales_age['Amount'].min(), sales_age['Amount'].max())
colors = plt.cm.viridis(norm(sales_age['Amount']))  # Using 'viridis' colormap

# Creating a bar plot without using the palette argument to avoid the warning
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, dodge=False)

# Manually setting the colors for the bars
for i, bar in enumerate(ax.patches):
    bar.set_color(colors[i])

# Adding grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels and title to the plot for better understanding
plt.title('Total Amount vs Age Group', fontsize=16)  # Title of the plot
plt.xlabel('Age Group', fontsize=12)  # X-axis label
plt.ylabel('Total Amount', fontsize=12)  # Y-axis label

# Rotating the x-axis labels for readability if age group names are long
plt.xticks(rotation=45)

# Displaying the plot
plt.show()

```

# From above graph we can see that most of buyers are of age group between 26-35 yrs female.

## State

```python
# Total no of order from top 10 states

# Grouping the data by 'State' and summing up the 'Orders', then selecting the top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (15, 5)})

# Creating a bar plot with a colorful palette

sns.barplot(data=sales_state, x='State', y='Orders', palette='plasma')

# Adding labels and title to the plot for better understanding
plt.title('Total Number of Orders from Top 10 States', fontsize=16)  # Title of the plot
plt.xlabel('State', fontsize=12)  # X-axis label
plt.ylabel('Number of Orders', fontsize=12)  # Y-axis label

# Rotating the x-axis labels for readability if state names are long
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

```python
# Total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (15, 5)})

# Creating a list of colors for each state (customize these colors as needed)
colors = sns.color_palette("Set2", n_colors=len(sales_state))  # Using the 'Set2' palette

# Creating a bar plot with custom colors
sns.barplot(data=sales_state, x='State', y='Amount', palette=colors)

# Adding grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels and title to the plot for better understanding
plt.title('Total Amount/Sales from Top 10 States', fontsize=16)  # Title of the plot
plt.xlabel('State', fontsize=12)  # X-axis label
plt.ylabel('Total Amount', fontsize=12)  # Y-axis label

# Rotating the x-axis labels for readability if state names are long
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

From the above graph web can see that most of the orders and total sales amount are from Uttar Pradrsh, Maharastra and Karnataka respectively.

## Marital Status

```python
# Creating a count plot for 'Marital_Status'
ax = sns.countplot(data=df, x="Marital_Status", palette=['#FF6347', '#4682B4'])  # Custom colors for the two bars

# Setting the figure size
sns.set(rc={'figure.figsize': (7, 5)})

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

```python
# Grouping the data by 'Marital_Status' and 'Gender', then calculating the total amount
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize':(8,6)})  # Slightly increased the figure size for better clarity

# Creating a bar plot using Seaborn with clear labels and a title
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender', palette='coolwarm')

# Adding labels and a title for better understanding
plt.title('Total Sales Amount by Marital Status and Gender', fontsize=16)  # Title of the plot
plt.xlabel('Marital Status', fontsize=12)  # X-axis label
plt.ylabel('Total Sales Amount', fontsize=12)  # Y-axis label

# Rotating the X-axis labels if necessary, for better readability (especially if labels are long)
plt.xticks(rotation=0)

# Displaying the plot
plt.show()
```

From above graph we can see that most of the buyer are married(Women)and the have high purchasing power.

### Product Category 


```python
# Setting the figure size
sns.set(rc={'figure.figsize': (25, 5)})

# Creating a countplot for 'Product_Category'
ax = sns.countplot(data=df, x='Product_Category', palette='Set2')  # Using a colorful palette

# Adding bar labels
for bars in ax.containers:
    ax.bar_label(bars, fontsize=12)  # Adjusting label font size

# Rotating the x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate labels, align right, adjust font size

# Adding title and labels
plt.title('Product Category Count', fontsize=16)  # Adding title
plt.xlabel('Product Category', fontsize=14)  # X-axis label
plt.ylabel('Count', fontsize=14)  # Y-axis label

# Adding grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= sales_state, x= 'Product_Category', y='Amount')
```

From above Graph we can see that most of sold product are from Food, Clothing, And Electronic category.

```python
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= sales_state, x='Product_ID', y= 'Orders')
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it contains 'Product_ID' and 'Orders' columns
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Create a figure and axis
fig1, ax1 = plt.subplots(figsize=(20, 7))  # Use plt.subplots() instead of plt.subplot()

# Create the bar plot
top_products.plot(kind='bar', ax=ax1)

# Set titles and labels
ax1.set_title('Top 10 Most Sold Products', fontsize=16)
ax1.set_xlabel('Product ID', fontsize=14)
ax1.set_ylabel('Total Orders', fontsize=14)
ax1.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
